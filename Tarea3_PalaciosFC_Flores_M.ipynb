>1. Resolver la siguiente ecuación recursiva usando funciones como se vió en clase
$$ D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2} \quad n\ge 3$$
con $D_1=0$ y $D_2 = 1$

### Solución 1 Flavio Palacios

```python
import numpy as np
def recur(d1:'number in the first position',d2:'number in the second position',n:'number to calculate'):
    """"
    Function to calculate the dn term given of the next recursive fucntion given
    the two initial numbers
    D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2}  for n>= 3
    """
    d = np.zeros(n)
    d[0] = d1
    d[1] = d2 
    def vector(i):
        nonlocal d
        d[i-1] = (i-1)*(d[i-2]+d[i-3])
    [vector(i) for i in range(3,n+1)]
    return d[-1]
recur(0,1,5)
```

### Solución 1 Moisés Flores

```python
##### Ejercicio 1
import numpy as np
def recursive_method1(n:"Número de elementos"):
    '''
    Esta es la programación de la ecuación recursiva  D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2} \quad n\ge 3
    con D_1=0 y D_2 = 1
    ----------
    N:'cantidad de términos'
    '''
    D=np.zeros(n)
    D[0] = 0; D[1] = 1;
    def llenar_vector(i):
        nonlocal D
        D[i-1] = (i-1)*D[i-2]+(i-1)*D[i-3]
    [llenar_vector(i) for i in range(3,n+1)]
    return np.array(D)
recursive_method1(6)
```

>6. Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9). 

### Solución 6 Flavio Palacios

```python
num = np.arange(2,10)
arr = np.arange(1,1001)
comp = lambda x,i: x if x % i == 0 else None
v = [np.array(list(filter(lambda x: x is not None,([comp(element,number) for element in arr])))) for number in num]
v
```
